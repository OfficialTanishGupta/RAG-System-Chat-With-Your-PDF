```python
import os
import logging

# 1. Clean up the output
os.environ["HF_HUB_DISABLE_SYMLINKS_WARNING"] = "1"
logging.getLogger("transformers").setLevel(logging.ERROR)

# 2. Import your modules
from src.pdf_loader import load_pdf
from src.chunking import split_text
from src.embeddings import create_embeddings
from src.retriever import create_vector_store, retrieve
from src.qa_system import answer_question
from sentence_transformers import SentenceTransformer

# 3. Load the model
print("Loading model... this may take a minute if downloading for the first time.")
model = SentenceTransformer("all-MiniLM-L6-v2")
print("Model loaded successfully!")
```

```python
text = load_pdf("../data/v8_engine_guide.pdf")
```

```python
chunks = split_text(text)
```

```python
embeddings = create_embeddings(chunks)
```

```python
index = create_vector_store(embeddings)
```

```python
question = "What is the need of Crankshaft?"

query_embedding = model.encode([question])
```

```python
results = retrieve(query_embedding, index, chunks)

context = " ".join(results)
```

```python
results = retrieve(query_embedding, index, chunks)

print(results)
```

```python
context = " ".join(results)

print(context)
```

```python
answer = answer_question(context, question)

print(answer)
```
